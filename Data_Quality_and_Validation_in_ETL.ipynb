**Question 1: Define Data Quality in the context of ETL pipelines. Why is it more than just data cleaning?**

Answer:

Data Quality in ETL pipelines refers to how fit the data is for its intended use across multiple dimensions.

Six Dimensions of Data Quality:

Accuracy – Data correctly represents real-world values
Completeness – No missing required data
Consistency – Uniform across systems
Validity – Follows formats and business rules
Timeliness – Up-to-date data
Uniqueness – No duplicates

Why it's more than just data cleaning:

Data cleaning is only one part. Data Quality also includes:

Validation (rules and constraints)
Standardization (format consistency)
Enrichment (adding value)
Monitoring (tracking quality)
Governance (policies & ownership)
Lineage (tracking data origin)

Data cleaning is reactive, while Data Quality is proactive.



**Question 2: Explain why poor data quality leads to misleading dashboards and incorrect decisions.**

Answer:

Poor data quality causes major business issues:

1. Garbage In, Garbage Out (GIGO)

Wrong data → wrong KPIs
Example: Duplicate sales inflate revenue

2. Misleading Trends

Missing data → fake decline
Inconsistent formats split metrics

3. Loss of Trust

One wrong metric → all reports doubted
Teams stop trusting data

4. Financial Impact

Overstock or stockouts
Wrong marketing targeting

5. Operational Failures

Invalid IDs → failed transactions
Duplicate orders → multiple shipments

Example:
A company saw 25% growth due to duplicate data, but actual growth was 8%, causing huge losses.

**Question 3: What is duplicate data? Explain three causes in ETL pipelines.**

Answer:

Duplicate data means multiple records represent the same real-world entity.

Causes:

1. Source System Issues

Multiple inserts
Retry logic
User mistakes

2. ETL Failures

Overlapping incremental loads
Missing deduplication logic
Job reruns after failure

3. Multiple Data Sources

Same entity in different systems
No Master Data Management
Question 4: Differentiate between exact, partial, and fuzzy duplicates.

Answer:

Exact Duplicates

All fields identical
Detected using GROUP BY

Partial Duplicates

Key fields same, others differ
Detected using business keys

Fuzzy Duplicates

Slight differences (typos, format)
Detected using similarity algorithms
Question 5: Why should data validation be performed during transformation rather than after loading?

Answer:

Reasons:

Prevent bad data entering warehouse
Fail fast (detect early errors)
Easier debugging
Better performance
Maintain quality standards
Ensure referential integrity

Best Practice Flow:

Extract → Validate → Transform → Validate → Load → Reconcile

Question 6: Explain how business rules help in validating data accuracy. Give an example.

Answer:

Business rules ensure data follows real-world logic.

Types of Rules:

Range checks (e.g., Quantity > 0)
Cross-field validation
Date logic
Compliance rules
Business policies

Example Rules:

COD orders ≤ 50,000
Employee discount ≤ 40%
Free shipping if order ≥ 1000
Perishable delivery ≤ 2 days

These rules prevent errors, fraud, and inconsistencies.

Question 7: Write an SQL query to list all duplicate keys and their counts using the business key.

Answer:

SELECT
    Customer_ID,
    Product_ID,
    Txn_Date,
    Txn_Amount,
    COUNT(*) AS Duplicate_Count
FROM Sales_Transactions
GROUP BY
    Customer_ID,
    Product_ID,
    Txn_Date,
    Txn_Amount
HAVING COUNT(*) > 1
ORDER BY Duplicate_Count DESC;
